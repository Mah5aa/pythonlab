The correlation coefficient is used to measure the strength of the relationship between two variables.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
def Pearson_correlation(X,Y):
    corr=0
    if len(X)==len(Y):
        Sum_xy = sum((X-X.mean())*(Y-Y.mean()))
        Sum_x_squared = sum((X-X.mean())**2)
        Sum_y_squared = sum((Y-Y.mean())**2)       
        corr = Sum_xy / np.sqrt(Sum_x_squared * Sum_y_squared)
    return corr
```

```python
x = np.array([4, 5, 10, 4, 3, 11, 14 , 6, 10, 12])
y = np.array([20, 22, 24, 17, 15, 26, 27, 22, 24, 28])
z = np.array([1, 2, 7, 12, 14, 6, 9 , 1, 6, 7])
print(Pearson_correlation(x,y))  
print(Pearson_correlation(x,z))  
print(Pearson_correlation(y,z))  
print(Pearson_correlation(x,x))
```

```python
print(np.corrcoef(x, y))
```

```python
def load_data(path):
    df=pd.read_csv(path)
    y=df.iloc[:,-1]
    x=df.iloc[:,:-1]
    return x,y

path= 'E:/master99/7/python lecture/python files/iris.csv'
dfx,dfy = load_data(path)
dfx
```

```python
c = np.corrcoef(dfx['sepal_length'],dfx['sepal_width'])
print('Correlations between sepal_length and sepal_width\n',c)
```

```python
corr = dfx.corr(method = 'pearson')
corr
plt.figure(figsize=(10,8), dpi =80)
sns.heatmap(corr,annot=True,fmt=".2f", linewidth=.5)
plt.show()
```
